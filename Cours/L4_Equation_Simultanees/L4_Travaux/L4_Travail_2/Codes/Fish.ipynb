```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
url = 'https://uploads-ssl.webflow.com/629e460595fdd36617348189/62a0fd19b6742078eed59f47_fish.out.txt'
fish_df = pd.read_table(url, sep="\t")
```

```python
fish_df.info()
```

```python
fish_df.head()
```

```python
fish_df.describe()
```

```python
fish_df['lqty'] = np.log(fish_df['qty'])
fish_df['lprice'] = np.log(fish_df['pricelevel'])
```

```python
# ElasticitÃ©s pas MCO
```

```python
# Offre
ols_o = sm.OLS(fish_df['lqty'], sm.add_constant(fish_df[['lprice', 'stormy', 'mixed']]))
ols_o = ols_o.fit(cov_type='HC0')
print(ols_o.summary()) 
```

```python
# Demande
ols_d = sm.OLS(fish_df['lqty'], sm.add_constant(fish_df[['lprice', 'rainy', 'cold', 'day2',
                                                         'day3', 'day4']]))
ols_d = ols_d.fit(cov_type='HC0')
print(ols_d.summary()) 
```
